### Methods of Feature Scaling:
1) Min-Max Scaling:\
   x`= x-min(x)/max(x)-min(x)
2) Standard Scaling (Z-score Normalization):\
   x`=x-mean(x)/std(x)
3) Robust Scaling:\
   x`=x-median(x)/IQR(x)
4) Logrithmic Scaling:\
   log or log10 of value

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

df=pd.read_csv("Titanic-Dataset.csv")
```

```python
#Sample Data
data={'Value':[10,20,30,40,50]}
df=pd.DataFrame(data)
print(df.head())
```

```python
# Min Max Scaler
scaler=MinMaxScaler()
df['Scaled_Values']=scaler.fit_transform(df[['Value']])
df
```

```python
# Standard Scaler or Z-score normalization
scaler = StandardScaler()
df['Scaled_Standar_Scaler']=scaler.fit_transform(df[['Value']])
df
```

```python
# Robust Scaler
scaler = RobustScaler()
df['Scaled_Robust_Scaler']=scaler.fit_transform(df[['Value']])
df
```

```python
# Logrithmic Scaling
#Sample Data
data={'Value':[10,20,30,40,50]}
df=pd.DataFrame(data)
print(df.head())
```

```python
# Robust Scaler
df['Scaled_through_log']=np.log(df[['Value']])
df['Scaled_through_log2']=np.log2(df[['Value']])
df['Scaled_through_log10']=np.log10(df[['Value']])
df
```
